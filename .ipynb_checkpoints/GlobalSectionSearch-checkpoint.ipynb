```python
import numpy as np
import matplotlib.pyplot as plt

# definisikan fungsi
def f(x):
    return (x-1)**2

# definisikan batas pencarian
a = -5
b = 7

# tentukan jumlah iterasi
max_iter = 10

# tentukan toleransi error
tolerance = 1e-5

# hitung lebar interval pencarian
width = abs(b-a)

# hitung jumlah segmen yang dibagi
num_segments = int(np.ceil(width/tolerance))

# hitung increment untuk setiap segmen
increment = width/num_segments

# inisialisasi nilai minimum awal
min_x = a
min_y = f(min_x)

# loop untuk setiap segmen
for i in range(num_segments):
    # hitung nilai x pada titik awal segmen
    x1 = a + i*increment
    # hitung nilai x pada titik akhir segmen
    x2 = x1 + increment
    # cari nilai minimum pada segmen ini
    for j in range(max_iter):
        # hitung nilai y pada kedua titik segmen
        y1 = f(x1)
        y2 = f(x2)
        # jika nilai minimum pada segmen ini ditemukan, update nilai minimum global
        if y1 < min_y:
            min_x = x1
            min_y = y1
        if y2 < min_y:
            min_x = x2
            min_y = y2
        # hitung nilai x pada titik tengah segmen
        xm = (x1 + x2)/2
        # hitung nilai y pada titik tengah segmen
        ym = f(xm)
        # jika nilai y pada titik tengah lebih kecil daripada kedua ujung segmen, cari minimum pada segmen ini
        if ym < y1 and ym < y2:
            if ym < min_y:
                min_x = xm
                min_y = ym
            break
        # update nilai x pada titik awal dan akhir segmen
        if y1 < y2:
            x2 = xm
        else:
            x1 = xm

# tampilkan hasil
print("Minimum ditemukan pada x = {:.4f} dengan nilai y = {:.4f}".format(min_x, min_y))
print("")

# plot fungsi dan titik minimum
x = np.linspace(a, b, 1000)
y = f(x)
plt.plot(x, y, label="Fungsi")
plt.plot(min_x, min_y, 'ro', label="Minimum")
plt.legend()
plt.grid()
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# define the function
x = sp.symbols('x')
f = (x-1)**2

# find the derivative of the function
f_prime = sp.diff(f, x)

# solve for critical points by setting f_prime = 0
critical_points = sp.solve(f_prime, x)

# evaluate the function at critical points and at endpoints of interval
values = [f.subs(x, cp) for cp in critical_points]
values += [f.subs(x, 0), f.subs(x, 2)]

# find the minimum value and the corresponding x
min_val = min(values)
min_x = [cp for cp in critical_points if f.subs(x, cp) == min_val][0]

# max_val = max(values)
# max_x = [cp for cp in critical_points if f.subs(x, cp) == max_val][0]

# print the results
print("Critical points: ", critical_points)
print("Function values at critical points and endpoints: ", values)
print("Minimum value: ", min_val)
# print("Maximum value: ", max_val)
# print("X value at Maximum: ", max_x)

# plot the function and the minimum value
x_vals = np.linspace(-5, 7, 100)
y_vals = [f.subs(x, xv) for xv in x_vals]
plt.plot(x_vals, y_vals, label="Function")
plt.plot(min_x, min_val, 'ro', label="Minimum")
plt.legend()
plt.grid()
plt.show()
```

```python
def middle(a,b,c):
    # first let's try with an x in the middle of the largest segment
    if b is None :
        return (a+c)/2
    if b-a > c-b :
        return (a+b)/2,True
    else :
        return (b+c)/2,False
min_err = 1e-5
def find_min(f,a,c,b=None,next_x=middle,err=min_err,max_iter=1e3,return_list=False):
    # find min with the golden section method
    if b is None:
        b = next_x(a,b,c)
    f_a= f(a)
    f_b = f(b)
    f_c = f(c)
    assert a<b
    assert f_a>=f_b
    assert f_c>=f_b
    n_iter = 0
    list_b = [a,b,c]
    while c-a > err and n_iter < max_iter:
        x,on_left = next_x(a,b,c)
        f_x = f(x)         
        if f_x < f_b :
            b = x
            f_b = f_x
            list_b.append(b)
        elif on_left :
            a = x
            f_a = f_x
        else :
            c = x
            f_c = f_x
        n_iter += 1
    if return_list:
        return list_b
    return b

print(find_min(f,0,10,return_list=True))
```

```python
x = [x for x in np.linspace(0,10,100)]
y = [f(x) for x in x]
plt.figure(figsize=(12,8))
minimum = min(y)
index = y.index(minimum)
plt.plot(x,y,label="function to optimised")
for i,c in enumerate(find_min(f,0,10,return_list=True)):
    plt.plot([c],[f(c)],marker="o",color="k",label="search point" if i == 0 else None,linestyle="None")   
plt.plot([x[index]],[y[index]],marker="o",color="r",label="objective point",linestyle="None")
plt.legend()
plt.grid()
plt.title(f'The midpoint finds the approximate solution to {min_err} in : {len(find_min(f,0,10,return_list=True))} steps in :( x  {x[index]:0.1f}  0.1 , y  {y[index]:0.1f} )')
plt.show()
```

```python
# Now let's use the golden ratio
phi = 1.618
def golden(a,b,c):
    # first let's try with an x in the middle of the largest segment
    if b is None :
        return a + (c-a)/(1+phi)
    if b-a > c-b :
        return b-(b-a)/(1+phi),True
    else :
        return b+(c-b)/(1+phi),False
```

```python
x = [x for x in np.linspace(0,10,100)]
y = [f(x) for x in x]
plt.figure(figsize=(12,8))
minimum = min(y)
index = y.index(minimum)
plt.plot(x,y,label="function to optimised")
for i,c in enumerate(find_min(f,0,10,return_list=True,next_x=golden)):
    plt.plot([c],[f(c)],marker="o",color="k",label="search point" if i == 0 else None,linestyle="None")   
plt.plot([x[index]],[y[index]],marker="o",color="r",label="objective point",linestyle="None")
plt.legend()
plt.grid()
plt.title(f'The golden ratio finds the approximate solution to {min_err} in : {len(find_min(f,0,10,return_list=True,next_x=golden))} steps')
plt.show()
```
